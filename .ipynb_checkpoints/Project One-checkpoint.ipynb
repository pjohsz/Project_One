```python
# Save path to movies data file in a variable
Nbr_data_file = "Resources/TheNumbersGross.csv"

# Use Pandas to read data
gross_data_df = pd.read_csv(Nbr_data_file)
gross_data_df.head()
```

```python
# Save path to movies data file in a variable
data_file = "Resources/movie_data.csv"

# Use Pandas to read data from IMDB
movie_df = pd.read_csv(data_file)
movie_df.head()
```

```python
# Clean the data frame to remove the "Â" value in the movie title
%timeit movie_df.movie_title.apply(lambda x: x.replace('Â',''))

movie_df.head()
```

```python
# remove rows where any of the columns are set to NaN
movie_df = movie_df.dropna(how='any')
movie_df.head()
```

```python
# Extract movie name, release date, budget and gross values
reduced_movie_df = movie_df.loc[:, ["movie_title", "budget", "gross","title_year"]]
reduced_movie_df

reduced_movie_df.head()
```

```python
# Extract the Star Wars Movie information
starwars_df = reduced_movie_df[reduced_movie_df['movie_title'].str.contains("Star Wars")] 
starwars_df

starwars_df.head(15)
```

```python
# Extract the Star Wars Movie information
startrek_df = reduced_movie_df[reduced_movie_df['movie_title'].str.contains("Star Trek")] 
startrek_df

startrek_df.head(15)
```

```python
# Get Starwars gross data for all Starwars films
starwars_gross_df = gross_data_df[gross_data_df['Franchise'].str.contains("Star Wars")] 
starwars_gross_df

starwars_gross_df.head(15)
```

```python
# Get Starwars gross data for all Starwars films
startrek_gross_df = gross_data_df[gross_data_df['Franchise'].str.contains("Star Trek")] 
startrek_gross_df

startrek_gross_df.head(15)
```

```python
# Sum the worldwide gross for all Starwars films
SW_Sum_World_Gross = starwars_gross_df.groupby('Franchise')['Worldwide_Gross'].sum()
SW_Sum_World_Gross

print(SW_Sum_World_Gross)
```

```python
# Sum the worldwide gross for all Starwars films
SW_Sum_Domestic_Gross = starwars_gross_df.groupby('Franchise')['Domestic_Gross'].sum()
SW_Sum_Domestic_Gross

print(SW_Sum_Domestic_Gross)
```

```python
# Sum the worldwide gross for all Starwars films
ST_Sum_World_Gross = startrek_gross_df.groupby('Franchise')['Worldwide_Gross'].sum()
ST_Sum_World_Gross

print(ST_Sum_World_Gross)
```

```python
# Sum the worldwide gross for all Starwars films
ST_Sum_Domestic_Gross = startrek_gross_df.groupby('Franchise')['Domestic_Gross'].sum()
ST_Sum_Domestic_Gross

print(ST_Sum_Domestic_Gross)
```

```python
# Save the Item Name and count into a summary data frame
SW_Fin_Sum_df = pd.DataFrame({"Franchise": "Star Wars",
                              "World Gross": SW_Sum_World_Gross,
                              "Domestic Gross": SW_Sum_Domestic_Gross})
# print summary table

SW_Fin_Sum_df.head()
```

```python
# Save the Item Name and count into a summary data frame
ST_Fin_Sum_df = pd.DataFrame({"Franchise": "Star Trek",
                              "World Gross": ST_Sum_World_Gross,
                              "Domestic Gross": ST_Sum_Domestic_Gross})
# print summary table

ST_Fin_Sum_df.head()
```
